```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import h5py
```

```python
data = pd.read_csv("penguins.csv")
data.head(10)
```

Filter out missing values

```python
print(data.shape)
data.dropna(inplace=True)
print(data.shape)
```

We are going to classify Species $(y)$ based on 2 features: Flipper Length and Body Mass $(X)$. The features should be normalised.

```python
X = data[["Flipper Length (mm)", "Body Mass (g)"]].to_numpy()
# encoding of categorical data in Species column
mapping = {"Adelie": 0, "Chinstrap": 1, "Gentoo": 2}
y = data["Species"].map(mapping).astype("int8").to_numpy()

# split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size = 0.3,
    stratify=y,
    random_state=42)

# scale features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
with h5py.File("train_data.h5", "w") as f:
    f.create_dataset("X", data=X_train)
    f.create_dataset("y", data=y_train)

with h5py.File("test_data.h5", "w") as f:
    f.create_dataset("X", data=X_test)
    f.create_dataset("y", data=y_test)
```
